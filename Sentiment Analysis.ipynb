# Chunking

```python
sample_text = inaugural.raw('1789-Washington.txt')
train_text = inaugural.raw('1793-Washington.txt')

custom_sent_tokenizer = PunktSentenceTokenizer(train_text)
tokenized = custom_sent_tokenizer.tokenize(sample_text)

def process():
    for i in tokenized:
        words = nltk.word_tokenize(i)
        tagged = nltk.pos_tag(words)
            
        Chunk = """cat:{<NN>}"""
        
        chunkParser = nltk.RegexpParser(Chunk)
        chunked = chunkParser.parse(tagged)
        
        print(chunked)
        
process()
```

# Combining Algos With a Vote

```python
from nltk.classify import ClassifierI
from statistics import mode
```

```python
class VoteClassifier(ClassifierI):
    def __init__(self,*classifiers):
        self._classifiers = classifiers
        
    def classify(self,features):
        votes = []
        for c in self._classifiers:
            v = c.classify(features)
            votes.append(v)
        return mode(votes)
    
    def confidence(self,features):
        votes = []
        for c in self._classifiers:
            v = c.classify(features)
            votes.append(v)
        choice_votes = votes.count(mode(votes))
        conf = choice_votes/len(votes)
        return conf
        
```

```python
voted_classifier = VoteClassifier(classifier,
                                  MNB_classifier,
                                  SGD_classifier)

print("voted classifier accuracy % : ", (nltk.classify.accuracy(voted_classifier,test_set))*100)
print("classifier : ", voted_classifier.classify(test_set[0][0]),"confidence : ", voted_classifier.confidence(test_set[0][0]))
```

```python
def sentiment(text):
    feats = find_features(word_tokenize(text))
    return voted_classifier.classify(feats),voted_classifier.confidence(feats)
```
